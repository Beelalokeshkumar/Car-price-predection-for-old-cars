```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('CarPrice_Assignment.csv')
df.head()
```

```python
df.drop('CarName',axis=1,inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df['car_ID'].unique()
```

```python
df.drop('car_ID',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['drivewheel'].unique()
```

```python
dw={'rwd':1,'fwd':2,'4wd':3}
df['drivewheel']=df['drivewheel'].map(dw)
df.head()
```

```python
df['drivewheel'].unique()
```

```python
df['doornumber'].unique()
```

```python
dn={'two':2,'four':4}
df['doornumber']=df['doornumber'].map(dn)
df.head()
```

```python
df['fueltype'].unique()
```

```python
ft={'gas':1,'diesel':2}
df['fueltype']=df['fueltype'].map(ft)
df.head()
```

```python
df['carbody'].unique()
```

```python
cb={'convertible':1, 'hatchback':2, 'sedan':3, 'wagon':4, 'hardtop':5}
df['carbody']=df['carbody'].map(cb)
df.head()
```

```python
df['aspiration'].unique()
```

```python
asp={'std':1, 'turbo':2}
df['aspiration']=df['aspiration'].map(asp)
df.head()
```

```python
df.drop('enginelocation',inplace=True,axis=1)
df.drop('enginesize',inplace=True,axis=1)
```

```python
df.head()
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df['cylindernumber']= label_encoder.fit_transform(df['cylindernumber'])
df['fuelsystem']= label_encoder.fit_transform(df['fuelsystem'])
df['wheelbase']= label_encoder.fit_transform(df['wheelbase'])
df['carlength']= label_encoder.fit_transform(df['carlength'])
df['carwidth']= label_encoder.fit_transform(df['carwidth'])
df['carheight']= label_encoder.fit_transform(df['carheight'])
df['enginetype']= label_encoder.fit_transform(df['enginetype'])
df['boreratio']= label_encoder.fit_transform(df['boreratio'])
df['stroke']= label_encoder.fit_transform(df['stroke'])
df['compressionratio']= label_encoder.fit_transform(df['compressionratio'])
df['price']= label_encoder.fit_transform(df['price'])
```

```python
df.head()
```

```python
df.info()
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(x_train,y_train)
```

```python
y_pred = svc.predict(x_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_pred,y_test)
```

```python
print(r2)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
mode=RandomForestRegressor().fit(x_train,y_train)
```

```python
y_pred1=mode.predict(x_test)
```

```python
print(r2_score(y_pred1,y_test))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression().fit(x_train,y_train)
```

```python
y_pred1=model.predict(x_test)
print(r2_score(y_pred1,y_test))
```

```python
from sklearn.decomposition import PCA
model=PCA(n_components =15)
x_train=model.fit_transform(x_train)
```

```python
x_test=model.transform(x_test)
```

```python
mode=RandomForestRegressor().fit(x_train,y_train)
y_pred1=mode.predict(x_test)
print(r2_score(y_pred1,y_test))
```
